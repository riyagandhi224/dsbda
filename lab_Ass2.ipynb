```python
def remove_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[column].clip(lower, upper)

numeric_cols = ["exercise_frequency", "mental_health_rating", "study_hours_per_day", "netflix_hours"]

for col in numeric_cols:
    df[col] = remove_outliers(col)


```
